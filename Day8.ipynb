```python
variables = pd.DataFrame({'normal': np.random.normal(size=15), 
                       'gamma': np.random.gamma(4, size=15), 
                       'poisson': np.random.poisson(size=15)})
print(variables)

variables.cumsum(1variables.cumsum(0).plot(subplots=True)).plot()
```

```python
variables.cumsum(0).plot(subplots=False)
variables.cumsum(0).plot(secondary_y='normal')
variables.cumsum(1).plot(secondary_y='gamma')
variables.cumsum(1).plot(secondary_y='poisson')
```

```python
variables.cumsum(1).plot(subplots=True)
variables.cumsum(0).plot(secondary_y='normal')
variables.cumsum(1).plot(secondary_y='gamma')
variables.cumsum(1).plot(secondary_y='poisson')
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(3, 15))
for i,var in enumerate(['normal','gamma','poisson']):
    variables[var].cumsum(0).plot(ax=axes[i], title=var)
axes[0].set_ylabel('cumulative sum')
```

```python
rng = np.random.RandomState(9)
x = rng.randn(1000)
y = rng.randn(1000)
colors = rng.rand(1000)
sizes = 2000 * rng.rand(1000)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.4,
            cmap='Blues')
plt.colorbar()
```
